## Module 1 — Python Basics

**What:** Variables, types, strings, and the f-string formatter — the alphabet of every later module.
**Why:** You can't analyse data you can't represent. Numbers, text, booleans, and lists are how every dataset enters Python.
**Key idea:** Python is dynamically typed — a name binds to *whatever* object you assign it to. Inspect with `type()` when in doubt.

```python
# Variables, types, expressions
name      = "Kader"          # str
age       = 38               # int
height_m  = 1.75             # float
is_active = True             # bool

print(type(name), type(age), type(height_m), type(is_active))

# Arithmetic + operator precedence
print(2 + 3 * 4)        # 14, not 20
print((2 + 3) * 4)      # 20
print(10 / 3, 10 // 3, 10 % 3, 2 ** 10)   # div, floor-div, mod, power
```

```python
# Strings: indexing, slicing, methods
s = "Data Science"
print(s[0], s[-1], s[5:12])           # 'D' 'e' 'Science'
print(s.upper(), s.lower(), s.replace("Science", "Engineering"))
print(s.split())                      # ['Data', 'Science']

# f-strings — the modern way to format
pi = 3.14159
print(f"{name} is {age} years old, height {height_m:.2f} m, π≈{pi:.3f}")
```

## Module 2 — Data Structures

**What:** Lists (ordered, mutable), Tuples (ordered, immutable), Dicts (key→value), Sets (unique, unordered).
**Why:** Each container has a job. Picking the right one decides whether your code is fast, readable, or buggy.
**Key idea:** Use **list** for sequences you'll change, **tuple** for fixed records, **dict** for lookups by key, **set** for membership/uniqueness.

```python
# Lists & tuples
nums = [3, 1, 4, 1, 5, 9, 2, 6]
nums.append(5)
print(sorted(nums))           # new sorted list
print(nums[2:5])              # slicing
nested = [[1,2],[3,4]]
print(nested[1][0])           # 3

point = (10.0, 20.5)          # tuple — can't be modified
x, y = point                  # unpacking
print(x, y)
```

```python
# Dictionaries — most-used structure in data work
person = {"name": "Kader", "age": 38, "skills": ["Python", "Excel"]}
person["city"] = "Riyadh"          # add
person["age"] += 1                  # update
print(person.get("email", "n/a"))   # safe lookup with default
for k, v in person.items():
    print(k, "->", v)
```

```python
# Sets — uniqueness and set algebra
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(a | b)   # union
print(a & b)   # intersection
print(a - b)   # difference
print(set("mississippi"))   # quick way to dedupe characters
```

## Module 3 — Programming Fundamentals

**What:** Branching (`if`), looping (`for`/`while`), reusable logic (`def`), error handling (`try/except`), and OOP (`class`).
**Why:** Data tasks are full of "for each row, if condition, do thing, handle the weird case." This module is that.
**Key idea:** Functions turn a block of logic into a single name you can reuse — the foundation of every library you'll later import.

```python
# Conditions + comparison + logical operators
def classify(n):
    if n < 0:
        return "negative"
    elif n == 0:
        return "zero"
    else:
        return "positive"

print([classify(x) for x in [-3, 0, 7]])   # list comprehension

# Logical operators: and / or / not
age, has_id = 22, True
print(age >= 18 and has_id)        # True
print(age < 13 or age > 65)        # False
print(not has_id)                  # False
```

```python
# Loops
total = 0
for i in range(1, 11):       # 1..10
    total += i
print("sum 1..10 =", total)

# while
n, fact = 5, 1
while n > 1:
    fact *= n
    n -= 1
print("5! =", fact)
```

```python
# Functions — parameters, defaults, return values, scope
def stats(values, ddof=0):
    """Return mean and population/sample variance."""
    mean = sum(values) / len(values)
    var  = sum((v-mean)**2 for v in values) / (len(values) - ddof)
    return mean, var

m, v = stats([2, 4, 4, 4, 5, 5, 7, 9])
print(f"mean={m}, var={v}")

# Variable scope — names inside a function are local unless declared global
counter = 0
def bump():
    global counter        # without this, assignment creates a local 'counter'
    counter += 1
bump(); bump(); print("counter:", counter)
```

```python
# Exception handling — never let a bad row crash a whole pipeline
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return float("inf")
    except TypeError as e:
        return f"bad input: {e}"

print(safe_divide(10, 2), safe_divide(10, 0), safe_divide(10, "x"))
```

```python
# Classes — bundle state + behaviour
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner   = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

    def __repr__(self):
        return f"BankAccount(owner={self.owner!r}, balance={self.balance})"

acc = BankAccount("Kader", 100)
acc.deposit(50)
print(acc)
```

## Module 1 — Intro to Matplotlib

**What:** Matplotlib is the foundational plotting library; nearly everything else (Pandas plots, Seaborn) wraps it.
**Why:** A two-line plot can reveal a pattern that hours of `.describe()` miss.
**Key idea:** Build a figure with `plt.subplots()`, draw on the returned `ax`, then `plt.show()`.

```python
import numpy as np, matplotlib.pyplot as plt
x = np.linspace(0, 4*np.pi, 200)
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(x, np.sin(x), label="sin")
ax.plot(x, np.cos(x), label="cos", linestyle="--")
ax.set(title="sin & cos", xlabel="x", ylabel="y"); ax.legend(); ax.grid(alpha=.3)
plt.show()
```

## Module 2 — Basic Visualization (bar, hist, pie, box, scatter, bubble)

**What:** Six chart types that answer the most common questions: comparisons, distributions, proportions, outliers, relationships.
**Key idea:** Pick the chart from the question, not the other way around. *Comparing categories?* bar. *Distribution?* hist or box. *Two numeric variables?* scatter.

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
rng = np.random.default_rng(0)
df = pd.DataFrame({
    "category": rng.choice(list("ABCD"), 200),
    "x": rng.normal(0, 1, 200),
    "y": rng.normal(0, 1, 200),
    "size": rng.integers(20, 200, 200),
})

fig, axes = plt.subplots(2, 3, figsize=(13, 7))

df["category"].value_counts().plot.bar(ax=axes[0,0], title="bar — counts")
df["x"].plot.hist(bins=20, ax=axes[0,1], title="hist — distribution of x")
df["category"].value_counts().plot.pie(ax=axes[0,2], autopct="%.0f%%", title="pie — share")
df.boxplot(column="x", by="category", ax=axes[1,0]); axes[1,0].set_title("box — x by category")
axes[1,1].scatter(df.x, df.y, alpha=.6); axes[1,1].set_title("scatter — x vs y")
axes[1,2].scatter(df.x, df.y, s=df["size"], alpha=.4); axes[1,2].set_title("bubble — size encodes 3rd var")

plt.tight_layout(); plt.show()
```

```python
# Area plot — like a line plot, but the area below is filled.
# Best for cumulative quantities or stacked parts of a whole.
import pandas as pd, numpy as np, matplotlib.pyplot as plt
years = pd.date_range("2018", periods=6, freq="YE").year
sales = pd.DataFrame({
    "Online": [10, 14, 22, 35, 45, 60],
    "Retail": [40, 38, 30, 25, 20, 18],
    "Wholesale": [25, 27, 30, 28, 32, 35],
}, index=years)
sales.plot.area(figsize=(8,3), title="Sales by channel (stacked area)")
plt.show()
```

```python
# Animation — show how a distribution evolves over time.
# In Colab the FuncAnimation renders as JS+HTML.
import numpy as np, matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

fig, ax = plt.subplots(figsize=(7,3))
x = np.linspace(0, 2*np.pi, 100)
line, = ax.plot(x, np.sin(x))
ax.set_ylim(-1.2, 1.2); ax.set_title("y = sin(x − t)")

def update(t):
    line.set_ydata(np.sin(x - t/5))
    return line,

ani = FuncAnimation(fig, update, frames=60, interval=80, blit=True)
plt.close(fig)
HTML(ani.to_jshtml())
```

```python
# Advanced Matplotlib customisation — anatomy of a publication-quality chart.
import numpy as np, matplotlib.pyplot as plt
x = np.linspace(0, 10, 200)
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, np.sin(x), label="sin", color="#1f77b4", linewidth=2)
ax.fill_between(x, np.sin(x), alpha=0.15, color="#1f77b4")
ax.axhline(0, color="grey", linewidth=.6)
ax.set(title="Anatomy of a chart", xlabel="time (s)", ylabel="amplitude")
ax.spines["top"].set_visible(False); ax.spines["right"].set_visible(False)
ax.legend(frameon=False); ax.grid(alpha=.25)
ax.annotate("peak", xy=(np.pi/2, 1), xytext=(2.5, 1.1),
            arrowprops=dict(arrowstyle="->"))
plt.show()
```

# PART 3 — Data Analysis with Python

We'll use the **classic auto-mpg-style dataset** (cars: mpg, weight, horsepower, etc.) end-to-end through every module so the steps connect.

```python
# Load the dataset once — used by all Part 3 modules
import pandas as pd, numpy as np
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
cols = ["mpg","cylinders","displacement","horsepower","weight",
        "acceleration","model_year","origin","car_name"]
cars = pd.read_csv(url, sep=r"\s+", names=cols, na_values="?")
print(cars.shape); cars.head()
```

## Module 1 — Importing Data Sets

**What:** Read CSV/Excel/SQL/web sources into a DataFrame and do a first inspection.
**Key idea:** Always start with `.shape`, `.dtypes`, `.head()`, `.info()`, `.describe()`. Five lines that prevent five hours of confusion.

```python
cars.info()
print(cars.describe(include="all").T.head(10))
```

## Module 2 — Data Wrangling

**What:** Fix what's broken — missing values, wrong types, weird scales, free-text categories.
**Why:** Models punish dirty data. Cleaning is unglamorous and usually >50% of the work.
**Key idea:** For each column ask: (1) right type? (2) any NaN? (3) needs scaling/encoding?

```python
# 1) Missing values
print("missing per column:\n", cars.isna().sum())

# Fill horsepower NaN with median (a defensible default)
cars["horsepower"] = cars["horsepower"].fillna(cars["horsepower"].median())

# 2) Type fixes
cars["origin"] = cars["origin"].map({1:"USA", 2:"Europe", 3:"Japan"}).astype("category")

# 3a) Normalisation (min-max scale weight to [0,1])
cars["weight_scaled"] = (cars.weight - cars.weight.min()) / (cars.weight.max() - cars.weight.min())

# 3b) Standardisation (z-score: mean 0, std 1) — preferred for most ML models
from sklearn.preprocessing import StandardScaler
cars["weight_std"] = StandardScaler().fit_transform(cars[["weight"]])

# 4) Binning continuous → categorical (e.g. mpg into Low/Med/High)
cars["mpg_bin"] = pd.cut(cars["mpg"], bins=[0, 18, 26, 50],
                          labels=["Low","Mid","High"])
print(cars["mpg_bin"].value_counts())

# 5) One-hot encoding of the categorical
cars_enc = pd.get_dummies(cars, columns=["origin"], drop_first=True)
cars_enc.head(3)
```

## Module 3 — Exploratory Data Analysis (EDA)

**What:** Look at the data from every angle before modelling. Distributions, correlations, group-bys.
**Key idea:** A correlation heatmap + a scatter matrix usually surface the 2–3 features that matter most.

```python
import seaborn as sns, matplotlib.pyplot as plt
num = cars.select_dtypes("number")
plt.figure(figsize=(8,5))
sns.heatmap(num.corr(), annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Correlation heatmap"); plt.show()
```

```python
# Group-by — does origin affect mpg?
print(cars.groupby("origin", observed=True)["mpg"].agg(["mean","median","count"]))

sns.boxplot(data=cars, x="origin", y="mpg"); plt.title("mpg by origin"); plt.show()
```

```python
# Pivot table — group-by on two dimensions at once.
# "Average mpg by origin × number of cylinders"
pivot = cars.pivot_table(values="mpg", index="origin",
                         columns="cylinders", aggfunc="mean", observed=True)
print(pivot.round(1))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Avg mpg by origin × cylinders"); plt.show()
```

## Module 4 — Model Development (Regression)

**What:** Build a predictive model — start with linear regression.
**Why:** Regression is the simplest, most-interpretable starting point; ~70% of business problems begin here.

Formula: $\hat{y} = b_0 + b_1 x_1 + b_2 x_2 + \dots$  — find the coefficients that minimise squared error.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score, mean_squared_error

features = ["cylinders","displacement","horsepower","weight","acceleration","model_year"]
X = cars[features].values
y = cars["mpg"].values

X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.2, random_state=0)

# 1) Simple linear regression on weight only
simple = LinearRegression().fit(X_tr[:, [3]], y_tr)
print(f"simple LR (weight only): R²={simple.score(X_te[:, [3]], y_te):.3f}")

# 2) Multiple linear regression
multi = LinearRegression().fit(X_tr, y_tr)
print(f"multiple LR (6 features): R²={multi.score(X_te, y_te):.3f}")
print("coefficients:", dict(zip(features, np.round(multi.coef_, 3))))

# 3) Polynomial regression in a Pipeline
poly = make_pipeline(PolynomialFeatures(degree=2, include_bias=False), LinearRegression())
poly.fit(X_tr, y_tr)
print(f"polynomial deg-2: R²={poly.score(X_te, y_te):.3f}")
```

## Module 5 — Model Evaluation & Refinement

**What:** Detect over/underfitting, validate honestly, regularise.
**Key idea:** A high score on training data is meaningless. Always score on **held-out data**, ideally via cross-validation.

Metric: $\mathrm{MSE} = \frac{1}{n}\sum (y_i - \hat{y}_i)^2$ &nbsp;·&nbsp; $R^2$ = fraction of variance explained.

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score, GridSearchCV

# Cross-validated R² for the multiple LR
cv = cross_val_score(LinearRegression(), X, y, cv=5, scoring="r2")
print(f"5-fold CV R²: mean={cv.mean():.3f} ± {cv.std():.3f}")

# Ridge — adds an L2 penalty to fight overfitting
grid = GridSearchCV(Ridge(), {"alpha":[0.01, 0.1, 1, 10, 100]}, cv=5, scoring="r2")
grid.fit(X, y)
print("best alpha:", grid.best_params_, "best R²:", round(grid.best_score_, 3))
```

```python
# Diagnostic: predicted vs actual
import matplotlib.pyplot as plt
y_pred = grid.predict(X_te)
plt.figure(figsize=(5,5))
plt.scatter(y_te, y_pred, alpha=.6)
plt.plot([y_te.min(), y_te.max()], [y_te.min(), y_te.max()], "r--")
plt.xlabel("actual mpg"); plt.ylabel("predicted mpg"); plt.title("Ridge — predictions vs truth")
plt.show()
print(f"test MSE: {mean_squared_error(y_te, y_pred):.2f}")
```

```python
# Overfitting vs underfitting — visualise it directly.
# Sweep polynomial degree and plot train R² vs validation R².
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import numpy as np, matplotlib.pyplot as plt

degrees, train_scores, val_scores = range(1, 8), [], []
for d in degrees:
    pipe = make_pipeline(PolynomialFeatures(d, include_bias=False), LinearRegression())
    pipe.fit(X_tr, y_tr)
    train_scores.append(pipe.score(X_tr, y_tr))
    val_scores.append(cross_val_score(pipe, X_tr, y_tr, cv=5, scoring="r2").mean())

plt.plot(degrees, train_scores, "o-", label="train R²")
plt.plot(degrees, val_scores,   "s-", label="validation R² (CV)")
plt.xlabel("polynomial degree"); plt.ylabel("R²")
plt.title("Underfit (left) → sweet spot → overfit (right)")
plt.axhline(0, color="grey", lw=.5); plt.legend(); plt.grid(alpha=.3); plt.show()
```
