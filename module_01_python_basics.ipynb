## 1. What is Python?

Python is a **high-level, interpreted, dynamically-typed** general-purpose language.

- *High-level* — you don't manage memory or pointers.
- *Interpreted* — code runs line-by-line; no separate compile step.
- *Dynamically typed* — a variable's type is decided by the value you assign to it, not declared upfront.

### Why data science chose Python
1. Huge scientific stack (NumPy, Pandas, scikit-learn, PyTorch).
2. Readable enough to use as pseudocode in a paper.
3. "Glue" language — connects to C, SQL, Spark, REST APIs effortlessly.

The cost: pure Python is **slow**. The libraries above sidestep this by running their hot loops in C/Fortran.

## 2. Jupyter Notebook basics

A notebook is a sequence of **cells**. Two types matter:

| Type | What it does | How you make one |
|---|---|---|
| **Code** | Runs Python; output appears below | default |
| **Markdown** | Renders formatted text (like this paragraph) | press `M` while a cell is selected |

### The kernel
A "kernel" is the Python process attached to your notebook. Variables you set in cell 3 are still alive in cell 7 *because the same kernel is running them*. Restart the kernel and they vanish — that's how you get a clean slate.

### Shortcuts that pay for themselves immediately
| Key | Action |
|---|---|
| `Shift + Enter` | run cell, move to next |
| `Ctrl/Cmd + Enter` | run cell, stay |
| `A` / `B` | new cell **A**bove / **B**elow (when cell is selected, not editing) |
| `D D` (twice) | delete cell |
| `M` / `Y` | switch to **M**arkdown / code (`Y`) |
| `Esc` / `Enter` | exit / enter edit mode |

### Magic commands
Special commands that start with `%` (line) or `%%` (whole cell). The most useful:

```python
# %time — how long did this line take?
%time sum(range(1_000_000))
```

```python
%%time
# %%time at the top of a cell times the whole cell
total = 0
for i in range(1_000_000):
    total += i
print(total)
```

## 3. Python syntax — the rules

### Indentation defines scope
Most languages use `{ }` to group statements. Python uses **indentation** (4 spaces by convention).
Mixing tabs and spaces or being inconsistent → `IndentationError`.

```python
# Correct indentation — 4 spaces inside the if-block
x = 10
if x > 0:
    print("positive")
    print("still inside the if")
print("outside the if")
```

### Comments
- `#` to end of line — single-line comment.
- Triple-quoted strings (`'''...'''` or `"""..."""`) are *not* comments technically — they're string literals — but they're commonly used for docstrings inside functions.

```python
# This is a single-line comment.
x = 5    # comments can also follow a statement.

"""
This triple-quoted string is a no-op at module level.
Used as a docstring at the top of a function/class/module.
"""
print(x)
```

### Statements vs expressions
- An **expression** evaluates to a value: `2 + 3`, `"hi".upper()`.
- A **statement** does something: `x = 5`, `if ...:`, `import ...`.
You can put a statement on its own line. Two statements on the same line need a `;` between them (but it's frowned upon).

## 4. Variables and expressions

A variable is a **name bound to a value**. Assignment uses `=`.

### Naming rules
- Letters, digits, underscores. Cannot start with a digit.
- Case-sensitive: `Age` and `age` are different.
- Convention (PEP 8): `snake_case` for variables and functions, `CapWords` for classes, `UPPER_SNAKE` for constants.

### Dynamic typing
You can rebind a name to a different type later — Python won't stop you.

```python
age = 38
name = "Kader"
height_m = 1.75
PI = 3.14159

print(name, age, height_m, PI)

# Rebind `age` to a string — legal but usually a smell.
age = "thirty-eight"
print(age, type(age))
```

```python
# Multiple assignment
a, b, c = 1, 2, 3
print(a, b, c)

# Swap two variables — no temp needed
x, y = 10, 20
x, y = y, x
print(x, y)
```

## 5. Numbers

Two main numeric types:
- `int` — whole numbers, **arbitrary precision** (Python ints can be huge).
- `float` — IEEE-754 double, ~15 significant digits, but with the usual binary-fraction quirks.

### Operators (in precedence order)
`**` (power) → unary `-` → `*  /  //  %` → `+  -`. Use parentheses when in doubt.

```python
# Arithmetic
print(7 + 3, 7 - 3, 7 * 3)
print(7 / 3)        # 2.333... — true division, ALWAYS float
print(7 // 3)       # 2 — floor division
print(7 % 3)        # 1 — modulo (remainder)
print(2 ** 10)      # 1024 — power
print(2 ** 100)     # arbitrary-precision int

# Underscores in numeric literals (purely cosmetic, helps readability)
million = 1_000_000
print(million)
```

```python
# The classic float pitfall — binary fractions can't represent 0.1 exactly.
print(0.1 + 0.2)              # 0.30000000000000004
print(0.1 + 0.2 == 0.3)       # False !

# Fix when comparing floats: use math.isclose
import math
print(math.isclose(0.1 + 0.2, 0.3))   # True
```

## 6. Booleans and None

- `bool` has exactly two values: `True` and `False` (capitalised).
- `None` is Python's "nothing here" singleton — its type is `NoneType`.

### Truthiness
Almost any value can be used in a boolean context. **Falsy** values: `False`, `0`, `0.0`, `""`, `[]`, `{}`, `()`, `None`. Everything else is **truthy**.

```python
flag = True
empty = None

print(flag, type(flag))
print(empty, type(empty))

# Truthiness in action
for v in [0, 1, "", "x", [], [0], None, 3.14]:
    print(repr(v), "->", bool(v))
```

## 7. Strings — creating them

A string is a **sequence of characters**. Quotes can be single, double, or triple (for multi-line).

```python
a = 'single quotes'
b = "double quotes"
c = """triple quotes
let you span
multiple lines"""
print(a); print(b); print(c)

# Escape sequences
print("tab\there")          # \t = tab
print("newline\nhere")      # \n = newline
print("a backslash: \\")   # \\ = literal backslash
print("a quote: \"")        # \" = literal "

# Raw strings — escape sequences are NOT processed.
# Useful for Windows paths and regex.
print(r"C:\Users\kader\file.txt")
```

## 7b. Strings — indexing and slicing

Strings are **0-indexed**. Negative indices count from the end (`-1` is the last char).
Slicing syntax: `s[start:stop:step]`. `stop` is exclusive.

```python
s = "Data Science"
#    0123456789...

print(s[0])         # 'D'
print(s[-1])        # 'e'
print(s[5:12])      # 'Science'   — indices 5..11
print(s[:4])        # 'Data'      — start defaults to 0
print(s[5:])        # 'Science'   — stop defaults to len
print(s[::2])       # 'Dt cec'    — every 2nd char
print(s[::-1])      # 'ecneicS ataD' — reversed

# len() gives the length
print(len(s))
```

## 7c. Strings — methods

Strings are **immutable** — methods return *new* strings. Common ones:

```python
s = "  Data Science  "

print(s.strip())                   # remove whitespace at the ends
print(s.upper(), s.lower())        # case
print(s.replace("Data", "Real"))
print(s.split())                   # ['Data', 'Science'] — splits on whitespace
print("-".join(["2024", "12", "31"]))  # '2024-12-31'
print("Data" in s, "X" in s)       # membership
print(s.find("Sci"), s.find("xx")) # index, or -1 if absent
print("data".startswith("da"), "data".endswith("ta"))
```

### Common pitfall — strings are immutable
Trying to assign to a character raises a `TypeError`. Build a new string instead.

```python
s = "hello"
# s[0] = "H"   # TypeError: 'str' object does not support item assignment
s = "H" + s[1:]
print(s)
```

## 8. Format strings (the modern way)

Three styles you will encounter, in order of preference:

| Style | Example | When to use |
|---|---|---|
| **f-string** (3.6+) | `f"{name} is {age}"` | always, in modern code |
| `.format()` | `"{} is {}".format(name, age)` | older codebases |
| `%`-formatting | `"%s is %d" % (name, age)` | very old code, logging configs |

### Format specs (the `:.2f` part inside `{ }`)
Inside an f-string `{ }` you can append `:fmt` to control width, precision, alignment, padding.

```python
name, age, pi = "Kader", 38, 3.14159265

# f-string — the default in modern Python
print(f"{name} is {age} years old.")
print(f"π is approximately {pi:.4f}")        # 4 decimal places
print(f"|{name:>10}|")                       # right-align in width 10
print(f"|{name:<10}|")                       # left-align
print(f"|{name:^10}|")                       # center
print(f"{1234567:,}")                        # 1,234,567  (thousands sep)
print(f"{0.045:.1%}")                        # 4.5% (percent format)
print(f"{255:08b} {255:#x}")                 # binary (padded), hex

# Show the variable AND its value (Py 3.8+) — handy for debugging
print(f"{age=}")                             # age=38
```

```python
# .format() — pre-f-string syntax
print("{} is {}".format(name, age))
print("{0} {0} {1}".format("ha", "!"))      # positional indices

# % formatting — legacy but still seen
print("%s is %d years old" % (name, age))
```

## 9. Type conversion (casting)

Use the type's name as a function: `int(x)`, `float(x)`, `str(x)`, `bool(x)`.

Key gotchas:
- `int("3.5")` is a `ValueError` — to go from string-of-float to int, do `int(float("3.5"))`.
- `bool(0) == False`; everything non-zero / non-empty is `True`.

```python
print(int("42"), type(int("42")))
print(float("3.14"))
print(str(99) + "%")        # join number with text
print(int(3.9))             # 3 — truncates toward zero (NOT round)

# When parsing fails, you get an error you should handle:
# int("abc")   # ValueError

# Combining with input()
# Note: input() always returns a string. Cast it to use as a number.
# n = int(input("Enter age: "))
```

## 11. Comparisons and boolean logic

### Comparison operators
`==`  `!=`  `<`  `>`  `<=`  `>=`

### Logical operators
`and`, `or`, `not` (lowercase, NOT `&&` / `||`).

### `==` vs `is`
- `==` compares **values**.
- `is` compares **identity** (same object in memory). Almost always use `==`. The one place `is` is right: comparing to `None` (`x is None`).

### Chained comparisons
`a < b < c` works the way it reads — that's a Python feature, not a coincidence.

```python
age = 22
print(age >= 18 and age < 65)
print(18 <= age < 65)               # equivalent — chained comparison
print(not (age < 18))

# is vs ==
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)        # True  — same contents
print(a is b)        # False — different list objects
print(a is a)        # True

# The one place `is` is correct
x = None
print(x is None)     # canonical None check
```

## 12. Reading a Python traceback (basic debugging)

When something breaks, Python prints a **traceback**. Read it from the bottom up — the last line tells you *what* went wrong; the lines above tell you *where*.

The five errors you'll meet first:

| Error | Cause | Fix |
|---|---|---|
| `NameError` | used a variable that doesn't exist | typo? defined it later? |
| `TypeError` | wrong type for the operation | cast or check `type(x)` |
| `ValueError` | right type, bad value (`int("abc")`) | validate before casting |
| `SyntaxError` | code can't be parsed | missing `:`, `)`, quote |
| `IndentationError` | mixed/misaligned spaces | use 4 spaces consistently |

```python
# Trigger and read each error. Uncomment one at a time.

# print(undefined_variable)        # NameError
# print("hi" + 5)                  # TypeError — can't add str + int
# int("abc")                       # ValueError
# eval("if True print('x')")       # SyntaxError (parsed at eval time)

# Print debugging — the simplest tool that actually works
def avg(values):
    print(f"DEBUG: values={values}, len={len(values)}")
    return sum(values) / len(values)

print(avg([3, 5, 7, 9]))
```
